```python
import pandas as pd
```

```python
df = pd.read_csv('agaricus-lepiota.csv')
```

```python
df.head()
```

```python
## remove all missing values
for column in df:
    df.drop(df[df[column] == '?'].index, inplace = True)
```

```python
print(f'new dataset size after removing missing values = {len(df)}')
```

```python
df.to_csv('mushroomDataset-cleaned.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## convert categorical data into numerical encoded values

from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
for column in df:
    df[f"{column}_encoded"] = ord_enc.fit_transform(df[[column]])
```

```python
encoded_df = df
```

```python
for col in encoded_df:
    if not 'encoded' in col:
        encoded_df = encoded_df.drop(col, axis=1)
```

```python
encoded_df.head()
```

```python
X = encoded_df.drop(['class_encoded'], axis=1)
y = encoded_df['class_encoded']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
f = open("encoded-py.txt", "w")
f.write('ENCODED VALUES \n\n')
for col in df:
    if not 'encoded' in col:
        unique = {}
        uniqueVal = df[col].unique()
        for val in uniqueVal:
            unique[val] = df[f'{col}_encoded'].loc[df[col] == val].values[0]
        f.write(str(col))
        f.write('\n')
        for k in unique:
            f.write(str(k))
            f.write(" ")
            f.write(str(unique[k]))
            f.write('\n')
        f.write('\n')
f.close()
```
